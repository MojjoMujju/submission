### Merging and Cleaning Dataset

Customer and Orders Dataset Merge

```python
# Merging Dataset
df_mergeCustOrders = df_cust.merge(
    df_orders,
    how='right',
    on='customer_id'
)
```

```python
df_mergeCustOrders['order_purchase_timestamp'] = pd.to_datetime(df_mergeCustOrders['order_purchase_timestamp'], format='%Y-%m-%d')
df_mergeCustOrders['order_approved_at'] = pd.to_datetime(df_mergeCustOrders['order_approved_at'], format='%Y-%m-%d')
df_mergeCustOrders['order_delivered_carrier_date'] = pd.to_datetime(df_mergeCustOrders['order_delivered_carrier_date'], format='%Y-%m-%d')
df_mergeCustOrders['order_delivered_customer_date'] = pd.to_datetime(df_mergeCustOrders['order_delivered_customer_date'], format='%Y-%m-%d')
df_mergeCustOrders['order_estimated_delivery_date'] = pd.to_datetime(df_mergeCustOrders['order_estimated_delivery_date'], format='%Y-%m-%d')
```

Customer, Orders, and Order Payments Dataset Merge

```python
df_mergeCustOrdersPayment = df_mergeCustOrders.merge(
    df_orderPayments,
    how='right',
    on='order_id'
)
```

## Visualization & Explanatory Analysis

Fungsi Untuk menampilkan Lineplot

```python
def lineplotfunct(dfmasking, title=str, freq=str, metode=str):
    if(metode == 'sum'):
      df_group_masking = dfmasking.groupby(
          by= pd.Grouper(key='order_approved_at', freq=freq))['payment_value'].sum()
    if(metode == 'mean'):
       df_group_masking = dfmasking.groupby(
          by= pd.Grouper(key='order_approved_at', freq=freq))['payment_value'].mean()

    fig, ax= plt.subplots(figsize=(18,8))
    plt.title(
        title,
        pad=18,
        fontdict=({'fontsize':18})
    )

    ax= (
        sns.lineplot(
            x= df_group_masking.index, y= df_group_masking,
            marker= '.',
            markersize= 10,
            markerfacecolor= 'red'
        ),
        ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1)),
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%b-%Y')),
        plt.xticks(rotation=45)
    )
```
