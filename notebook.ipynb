## Dataframe Masking Order_approved_at
This masking the goal is to get Year-Month-days from order_approved_at column to analysis the Frequency and Monetary by month

```python
dfmasking = df_mergeCustOrdersPayment[['order_approved_at', 'payment_value']].copy()
```

PLEASE READ THIS NOTES!!!

Notes:
- disini seharusnya order_approved_at lebih baik jangan di drop (tapi dapat dilakukan imputasi nilai null value), dikarenakan terdapat column payment_value yang berisikan nilai keuangan dari customer.
- akan tetapi disini hanya untuk uji coba, nilai null value akan di drop untuk dilakukan analisa lebih lanjut terhadap jumlah
dan rata-rata payment_value pada setiap bulan.

Suggestion:
- Tidak melakukan drop rows, yang memiliki dampak pada payment values. hal ini berhubungan dengan stackholder lain dalam menentukan keputusan seperti tim bisnis.

```python
dfmasking.dropna(subset=['order_approved_at'], inplace=True)
```

```python
dfmasking['order_approved_at'] = pd.to_datetime(dfmasking['order_approved_at'], format='%Y-%m-%d')
```

```python
dfmasking['year'] = dfmasking['order_approved_at'].dt.year.astype(int)
dfmasking['month'] = dfmasking['order_approved_at'].dt.month.astype(int)
dfmasking['days'] = dfmasking['order_approved_at'].dt.day.astype(int)
dfmasking['year_month'] = dfmasking['order_approved_at'].dt.to_period('M').astype(str)
```

```python
dfmasking.head(5)
```

```python
def lineplotfunct(title=str, freq=str, metode=str):
    if(metode == 'sum'):
      df_group_masking = dfmasking.groupby(
          by= pd.Grouper(key='order_approved_at', freq=freq))['payment_value'].sum()
    if(metode == 'mean'):
       df_group_masking = dfmasking.groupby(
          by= pd.Grouper(key='order_approved_at', freq=freq))['payment_value'].mean()

    fig, ax= plt.subplots(figsize=(18,8))
    plt.title(
        title,
        pad=18,
        fontdict=({'fontsize':18})
    )

    ax= (
        sns.lineplot(
            x= df_group_masking.index, y= df_group_masking,
            marker= '.',
            markersize= 10,
            markerfacecolor= 'red'
        ),
        ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1)),
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%b-%Y')),
        plt.xticks(rotation=45)
    )
```

```python
df_groupby_month_sum = dfmasking.groupby(['year_month'], as_index=False)['payment_value'].sum()
df_groupby_month_average = dfmasking.groupby(['year_month'], as_index=False)['payment_value'].mean()
```

# RFM (Recency, Frequency, Monetary)
RFM (Recency, Frequency, Monetary) adalah Teknik yang Sangat Sederhana yang dapat kita terapkan dengan sangat mudah dan mendapatkan analisis yang sangat berguna untuk Segmentasi Pelanggan kita.

RFM mempertimbangkan nilai recency, frequency, dan monetary untuk setiap pelanggan.

source: https://medium.com/@jackiekhuu.work/customer-segmentation-rfm-analysis-8007d62101cc

```python
# How to Get Customer Recency
recency = df_mergeCustOrdersPayment.groupby('customer_id')['order_approved_at'].max()
recency = max(recency) - recency
recency = recency.dt.days
#recency.head()
```

## Monetary - Visualization

### Monetary Total Jumlah Transaksi Pada Tiap Provinsi

```python
df_group_by_sum = df_mergeCustOrdersPayment.groupby(by='customer_state', as_index=False)['payment_value'].sum()
df_sort_values_sum = df_group_by_sum.sort_values('payment_value', ascending=False)
#df_sort_values_sum
```

```python
df_group_by_mean = df_mergeCustOrdersPayment.groupby(by='customer_state', as_index=False)['payment_value'].mean()
df_sort_values_mean = df_group_by_mean.sort_values('payment_value', ascending=False)
#df_sort_values_mean
```

```python
fig, ax = plt.subplots(figsize=(10,12), nrows=2, ncols=1)
ax= (
    sns.barplot(x= df_sort_values_sum['customer_state'], y= df_sort_values_sum['payment_value'], palette='pastel', ax=ax[0]),
    ax[0].set(title='Jumlah total Nilai Transaksi Setiap Provinsi', xlabel='Nama Provinsi', ylabel='Nilai Transaksi'),
    sns.barplot(x= df_sort_values_mean['customer_state'], y= df_sort_values_mean['payment_value'], palette='pastel', ax=ax[1]),
    ax[1].set(title='Rata-Rata total Nilai Transaksi Setiap Provinsi', xlabel='Nama Provinsi', ylabel='Nilai Transaksi')
)
```

Insight:
- SP merupakan nama provinsin dengan pembelian terbanyak, tetapi dengan rata-rata pengeluaran pembelian paling kecil, diantara kode nama provinsi lainnya.
- PB merupakan kode nama provinsi dengan jumlah rata-rata pembelian tertinggi, akan tetapi memiliki jumlah pembelian total yang kalah jauh dari SP.
